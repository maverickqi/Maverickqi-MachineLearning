```python
# 1、获取数据
import pandas as pd

data = pd.read_csv('/Users/zhangqihao/Desktop/PYTHON/机器学习/机器学习资料6/data.csv',encoding='gbk')
```

```python
# 2、划分数据
x_train = data[["2019年国际排名","2018世界杯","2015亚洲杯"]]
```

```python
# 3、MAX-MIN降维
from sklearn import preprocessing

transfer = preprocessing.MinMaxScaler()
x_train = transfer.fit_transform(x_train)
```

```python
# 4、预估器流程
from sklearn.cluster import KMeans

estimator = KMeans(n_clusters=5)
estimator.fit(x_train)
```

```python
# 5、合并聚类结果，插入到原数据中
y_predict = estimator.predict(x_train)

result = pd.concat((data,pd.DataFrame(y_predict)),axis=1)

result.rename({0:u'聚类'},axis=1,inplace=True)
result
```
